```julia
include("airplane_header.jl")
using Distributions
using Plots
```

```julia
num_sims=100000
SE_RJ=zeros(num_sims)
SE_NB=zeros(num_sims)
SE_WB=zeros(num_sims)

for n=1:num_sims
    SE_RJ[n]=SERP_Distributions(350,75,50000)
    SE_NB[n]=SERP_Distributions(500,150,100000)
    SE_WB[n]=SERP_Distributions(2000,300,250000)
end

SE=vcat(SE_RJ,SE_NB,SE_WB)
histogram(SE_RJ)
```

```julia
histogram(SE_RJ)
savefig("RJ_histogram.png")
```

```julia
histogram(SE_NB)
savefig("NB_histogram.png")
```

```julia
histogram(SE_WB)
savefig("WB_histogram.png")
```
